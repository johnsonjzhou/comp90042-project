## Load the datasets

```python
data_names = ["train-claims", "dev-claims", "evidence"]
train_claims, dev_claims, all_evidences \
    = load_as_dataframe(data_names, full_evidence=True)
```

```python
x = pd.concat([train_claims, dev_claims]).reset_index()[["evidences", "evidence_text"]].set_index("evidences")
x
```

```python
x = x.drop_duplicates()
x
```

```python
x = x.to_dict(orient="index")
x = {k:v["evidence_text"] for k,v in x.items()}
x
```

```python
with open("./data/train_dev_evidence.json", mode="w") as f:
    json.dump(x, f)
```
